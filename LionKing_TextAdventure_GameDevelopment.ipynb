#### Lion King Hangman, Lion King T/F Quiz and Lion King Quiz are all games based on the beloved Disney classic, Lion King. These three stages of games are an excellent way for fans of the Lion King movie to test and improve their knowledge about the characters and events of the movie, and also to have fun. 

#### In all three games, the player's score is displayed at the end and the player may have the option to play again if they do not reach the goal score of 30 points or pass the final quiz. Each stage of the game must be passed by reaching a score of 30 points, if the player fails to reach the goal score, they will have the option to retry the stage. This allows players to keep trying until they successfully pass the stage and advance to the next one.

## Stage 1

#### In Lion King Hangman, players are given descriptions of characters from the movie as a hint and must guess the names of the characters by either guessing letters or the full name. If a letter is correctly guessed, it will be revealed in the character's name and if the full name is correctly guessed, the player earns 10 points. The game continues with different characters until the player reaches a certain number of points or runs out of turns. 

## Stage 2

#### In Lion King T/F Quiz, players are presented with a series of true or false questions about the Lion King movie and must enter 't' for true or 'f' for false as their answer. Points are earned for correct answers. 

## Stage 3

#### In Lion King Final Quiz, players are presented with a series of multiple choice questions about the Lion King movie and must select the correct answer to earn points.

```python

# Function that allows the player to either retry stage 1 or move on to stage 2
def bridge1():
    
    # print instructions
    print("*"*40)
    print("""Do you want to retry Stage 1 or move on to Stage 2?
    1. Retry Stage 1
    2. Go to Stage 2""")
    
    #start a loop to get player's input
    while True:
        user_input = input("please select 1 or 2: ")
        #filter invalide input
        if user_input not in ["1","2"]:
            print("Invalid input. Please enter 1  or 2")
            continue
        else:
            break
    # based user input put the player into stage 1 or stage 2
    if int(user_input) == 1:
                  stage_1()
    else:
                  stage_2()
            
            
# Function that allows the player to either retry stage 2 or move on to stage 3         
def bridge2():
    
    # print instructions
    print("*"*40)
    print("""Do you want to retry Stage 2 or move on to Stage 3?
    1. Retry Stage 2
    2. Go to Stage 3""")
    
    #start a loop to get player's input
    while True:
        user_input = input("please select 1 or 2: ")
        #filter invalide input
        if user_input not in ["1","2"]:
            print("Invalid input. Please enter 1  or 2")
            continue
        else:
            break
    # based user input put the player into stage 2 or stage 3
    if int(user_input) == 1:
                  stage_2()
    else:
                  stage_3()
    

def stage_1():
    #question pool for stage 1 in a dictionary name,description,answer
    characters = [
        {"name":"Simba", "description":"The lion prince of the Pride Lands.", "answer":"simba"},
        {"name":"Mufasa", "description":"The king of the Pride Lands and Simba's father.", "answer":"mufasa"},
        {"name":"Scar", "description":"Mufasa's younger brother and the main antagonist of the film.", "answer":"scar"}
    ]
    
    score = 0 # Initialize the score to 0
    goal_score = 30 # The player needs to score 30 points to pass stage 1
    turns = 6 # The player has 6 turns to guess the correct character
    
    #welcome message
    print("*"*40)
    print("Welcome to Stage 1: Lion King Hangman game")
    print("*"*40)
    
    input("Press Enter to start the game: ")
    
    print("Hint: You can either guess a letter or guess the full name of the character.")
     # Main loop that runs the word game
    while score < goal_score:
        
        # Randomly select a character from the list
        character = random.choice(characters)
        
         # Assign the character's name, description and answer to variables of the same name
        name = character["name"]
        description = character["description"]
        answer = character["answer"]
        
        # a variable to store player's guesses
        guesses = ""
        
        # showing th description of the character
        print("*"*40)
        print("Description: ", description)
        print("*"*40)
        
        #start a main loop of the game
        while turns > 0:
            failed = 0
            # Check if the letter has been guessed
            for char in answer:
                if char in guesses: # If yes, print the letter
                    print(char, end=" ")
                else:
                    print("_", end=" ") # If no, print an underline
                    failed += 1 # Increment the failed variable
            if failed == 0: # If all letters are guessed correctly
                print("\nCongratulations! You won 10 points.")
                score += 10 # Add 10 points to the score
                break
                
            print("\nYou have", turns, "turns left")
            print("Your current score is", score)
            guess = input("\nEnter your guess: ").lower()
            # If the player's guess is equal to the correct answer print Correct
            if guess == answer:
                print("Correct!")
                score += 10 #adds 10 points to the player's score
                break
            # If the guess is not equal to the answer but is present in the answer, the code prints "Correct letter!"
            elif guess in answer:
                print("Correct letter!")
                guesses += guess
            # if the guess is wrong deduct 1 turns
            else:
                print("Wrong")
                turns -= 1
        # when player didn't reached the goal score of the stage ask play again
        if score < goal_score:
            # when player lose all there chances ask them playagain or quit the game
            if turns == 0:
                print("*"*40)
                print("Game over!\n")
                print("*"*40)
                play_again = input("Do you want to retry stage 1 again? (y/n)").lower()
                if play_again == 'y':
                    stage_1()
                else:
                    print("*"*40)
                    print("Thank you for playing! Your final score is",score)
                    print("*"*40)
                    break
        # when the player reached the goal score than show congrat message and break out from the stage
        else:
            print("*"*40)
            print("Congratulations! You passed the Stage 1 with a final score of",score)
            print("*"*40)
            break
            
            
def stage_2():
    #question pool for stage 2 in a dictionary question, answer
    questions = [
        {"question": "Is Simba the main character in The Lion King?", "answer": "t"},
        {"question": "Does Simba's father, Mufasa, die in the beginning of the movie?", "answer": "t"},
        {"question": "Does Simba return to the Pride Lands to take back his rightful place as king?", "answer": "t"},     
        {"question": "Does Simba's uncle Scar rule the Pride Lands for the entire movie?", "answer": "f"},
        {"question": "Is the Lion King a real story?", "answer": "f"},
        {"question": "Is Simba's mother still alive in the movie?", "answer": "f"}
        
    ]
    # Number of chances user has to retake the quiz
    chances = 5
    score = 0 # Initialize the score to 0
    total_score = 30 # The player needs to score 30 points to pass stage 2
    
    #welcome message
    print("*"*40)
    print("Welcome to the Stage2: T/F Lion King Quiz!")
    print("*"*40)
    
    input("Press Enter to start the game: ")

    # Main loop that runs the quiz
    
    while True:
        
        # randomize the questions order
        random.shuffle(questions)
        for i, question in enumerate(questions):
            # display the number of chances remaining
            print(f"Chances remaining: {chances}")
            user_answer = input(f"{i+1}. {question['question']} (T/F) ").lower()
            # handle invalid inputs
            if user_answer not in ['t', 'f']:
                print("Invalid input. Please enter 't' for true or 'f' for false.")
                continue
            if user_answer == question["answer"]:
                # award the same number of points for each correct answer
                print("\nCongratulations! You won 10 points.")
                score += 10
                
            else:
                print("Incorrect.")
                # decrement the chances if the user answers incorrectly
                chances-=1
            print(f"Current score: {score}")
            print(f"Total score: {total_score}")
            print("---"*5)
            if chances == 0:
                print("You failed")
                break
            elif score == total_score:
                break
            else:
                continue
        # when player didn't reached the goal score of the stage ask play again
        if score < total_score:
            # when player lose all there chances ask them playagain or quit the game
            if chances == 0:
                print("Game over!\n")
                play_again = input("Do you want to retry stage 2 again? (y/n) ")
                if play_again.lower() == "y":
                    stage_2()
                else:
                    print("*"*40)
                    print("Thank you for playing! Your final score is",score)
                    print("*"*40)
                    break
         # when the player reached the goal score than show congrat message and break out from the stage
        else:
            print("*"*40)
            print("Congratulations! You passed the Stage 2 with a final score of",score)
            print("*"*40)
            break
            
           

            
            
def stage_3():
    # Define the questions and answers with list and dictionary
    questions = [
        {
            "question": "Who is the protagonist of the Lion King?", 
            "choices": ["Mufasa", "Scar", "Simba", "Nala"],
            "answer": "Simba",
            "points": 10
        },
        {
            "question": "Who is Simba's best friend?", 
            "choices": ["Timon", "Pumbaa", "Rafiki", "Zazu"],
            "answer": "Pumbaa",
            "points": 10
        },
        {
            "question": "Who is the villain of the Lion King?", 
            "choices": ["Mufasa", "Scar", "Simba", "Nala"],
            "answer": "Scar",
            "points": 10
        },
        {
            "question": "What is the name of Simba's love interest?", 
            "choices": ["Sarabi", "Nala", "Rafiki", "Zazu"],
            "answer": "Nala",
            "points": 10
        }
    ]

    # Keep track of the number of correct answers
    correct_answers = 0 # Initialize the correct answers to 0
    total_score = 0 # Initialize the score to 0
    score_to_reach = 30 # The player needs to score 30 points to pass stage 3
    
    
    #welcome message
    print("*"*40)
    print("Welcome to Stage3: the Lion King Final Quiz!")
    print("*"*40)
    
    input("Press Enter to start the Final Quiz: ")
    
    while True:
    # Shuffle the order of the questions
        random.shuffle(questions)

        # Ask the questions
        for question in questions:
            print(question["question"])
            for i, choice in enumerate(question["choices"]):
                print(f"{i+1}. {choice}")
            response = input("Enter the number of your answer: ")

            # Check if the answer is correct
            if question["choices"][int(response)-1] == question["answer"]:
                print("Correct!\n")
                correct_answers += 1
                total_score += question["points"]
            else:
                print("Incorrect. The correct answer is:", question["answer"],"\n")
            print(f"Current score: {total_score}")
            print(f"You have answered {correct_answers} out of {len(questions)} questions correctly.\n\n")

        # Display the final score
        print("\n\n\n")

        # Display the final score
        print("\n\n\n")
        print("*"*40)
        print("\t\t\t\t\t Quiz Results")
        print("*"*40)
        # Display the final score using f print
        print(f"You got {correct_answers} out of {len(questions)} questions correct.")
        print(f"Your total score is {total_score}")
        
        # when player didn't reached the goal score of the stage ask play again
        if total_score < score_to_reach:
            print("Game over!\n")
            play_again = input("Do you want to retry stage_3 again? (y/n) ")
            
            if play_again.lower() == "y":
                stage_3()
            else:
                print("Thank you for playing!")
                break
        else:
            print("Congratulations! You passed the quiz!")
            break
        
        
        

import random
stage_1()
bridge1()
stage_2()
bridge2()
stage_3()
```
